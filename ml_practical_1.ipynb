```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data[:50,0]
y = iris.data[:50,1]

iris = plt.scatter(X,y)
```

Find m and c using matrix method

```python
X_reshaped = X.reshape(-1,1)
X_reshaped.shape
```

```python
y_reshaped = y.reshape(-1,1)
y_reshaped.shape
```

```python
X = np.hstack((np.ones_like(X_reshaped),X_reshaped))
X.shape
```

```python
# Calculate X^T * y
X_T_y = np.dot(X.T, y_reshaped)
```

```python
# Calculate X^T * X
X_T_X = np.dot(X.T, X)

# Calculate (X^T * X)^(-1)
X_T_X_inv = np.linalg.inv(X_T_X)

# Calculate coefficients B = (X^T * X)^(-1) * X^T * y
B = np.dot(X_T_X_inv, X_T_y)
```

```python
# B contains [c, m]
c = B[0]
m = B[1]
```

```python

print(f"Intercept (c): {c.item():.4f}")
print(f"Slope (m): {m.item():.4f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 1], y, label='Original Data') # X[:, 1] because X is the design matrix with a column of ones

# Generate points for the regression line
x_line = np.linspace(X[:, 1].min(), X[:, 1].max(), 100)
y_line = m * x_line + c
plt.plot(x_line, y_line, color='red', label=f'Regression Line (y = {m.item():.2f}x + {c.item():.2f})')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear Regression Best Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```
